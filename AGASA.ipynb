```
 
 def objective_value_t (weight, value, population, threshold):
   fitness = np.empty(population.shape[0])

   for i in range (population.shape[0]):
       S1 = np.sum(population[i]*value)
       S2 = np.sum(population[i]*weight)
       if S2 <= threshold:
         fitness[i] = S1
       else:
         fitness[i] = 0 
   avg_fitness = np.round(np.average(fitness))
   max_fitness = np.max(fitness)
   return fitness, avg_fitness,max_fitness
```

```
def objective_value(weight,value,population,threshold):
  fitness_sum = 0
  weight_sum = 0
  t=1
  for i in range(len(population)):
    fit_v = population[-t]*value[-t]
    fitness_sum = fitness_sum + fit_v
    fit_w = population[-t]*weight[-t]
    weight_sum = weight_sum + fit_w
    t = t+ 1
  
  if weight_sum <= threshold:
    return fitness_sum 
  else:
    return 0
```

```
def find_parents_ts (all_solutions):
  parents = np.empty((0,np.size(all_solutions,1)))

  for i in range(2):
    indices_list = np.random.choice(len(all_solutions),5,replace=False)
    
    posb_parent_1 = all_solutions[indices_list[0]]
    posb_parent_2 = all_solutions[indices_list[1]]
    posb_parent_3 = all_solutions[indices_list[2]]
    posb_parent_4 = all_solutions[indices_list[3]]
    posb_parent_5 = all_solutions[indices_list[4]]

    obj_func_parent_1 = objective_value(weight,value,posb_parent_1,knapsack_threshold)
    obj_func_parent_2 = objective_value(weight,value,posb_parent_2,knapsack_threshold)
    obj_func_parent_3 = objective_value(weight,value,posb_parent_3,knapsack_threshold)
    obj_func_parent_4 = objective_value(weight,value,posb_parent_4,knapsack_threshold)
    obj_func_parent_5 = objective_value(weight,value,posb_parent_5,knapsack_threshold)

    max_obj_func = max(obj_func_parent_1,obj_func_parent_2,obj_func_parent_3,obj_func_parent_4,obj_func_parent_5)

    if max_obj_func == obj_func_parent_1:
      selected_parent = posb_parent_1
    elif max_obj_func == obj_func_parent_2:
      selected_parent = posb_parent_2
    elif max_obj_func == obj_func_parent_3:
      selected_parent = posb_parent_3
    elif max_obj_func == obj_func_parent_4:
      selected_parent = posb_parent_4
    else:
      selected_parent = posb_parent_5
    
    parents = np.vstack((parents,selected_parent))

  parent_1 = parents[0,:]
  parent_2 = parents[1,:]

  return parent_1, parent_2
```

```
def mutation(child_1,child_2,prob_mutation):
  mutated_child_1 = np.empty((0,len(child_1)))
  obj_child1 = objective_value(weight,value,child_1,knapsack_threshold)
  obj_child2 = objective_value(weight,value,child_2,knapsack_threshold)
  avg_fitness = objective_value_t(weight,value,initial_population,knapsack_threshold)[1]
  max_fitness = objective_value_t(weight,value,initial_population,knapsack_threshold)[2]
  v =0 
  x = (max_fitness - obj_child1)/(max_fitness - avg_fitness)
  y = (max_fitness - obj_child2)/(max_fitness - avg_fitness)
  if obj_child1 > avg_fitness:
    for i in child_1:
      rand_num_to_mutate_or_not_1 = np.random.rand()

      if rand_num_to_mutate_or_not_1 < prob_mutation:
        if child_1[v] ==0:
          child_1[v]=1
        
        else:
          child_1[v] =0
        
        mutated_child_1 = child_1

        v=v+1
    
  else:
    if obj_child1 <= avg_fitness:
      for i in child_1:
        rand_num_to_mutate_or_not_1 = np.random.rand()
        if rand_num_to_mutate_or_not_1 < x*prob_mutation:
          if child_1[v] == 0:
            child_1[v] =1
          else:
            child_1[v]=0
          mutated_child_1 = child_1
          v = v+1

  mutated_child_2 = np.empty((0,len(child_2)))

  v =0 
  if obj_child2 > avg_fitness:
    for i in child_2:
      rand_num_to_mutate_or_not_2 = np.random.rand()

      if rand_num_to_mutate_or_not_2 < prob_mutation:
        if child_2[v] ==0:
          child_2[v]=1
        
        else:
          child_2[v] =0
        
        mutated_child_2 = child_2

        v=v+1
    
  else:
    if obj_child2 <= avg_fitness:
      for i in child_2:
        rand_num_to_mutate_or_not_2 = np.random.rand()
        if rand_num_to_mutate_or_not_2 < y*prob_mutation:
          if child_2[v] == 0:
            child_2[v] =1
          else:
            child_2[v]=0
          mutated_child_2 = child_2
          v = v+1
  

  return mutated_child_1, mutated_child_2
```
